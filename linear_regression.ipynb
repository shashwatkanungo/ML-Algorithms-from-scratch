```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## generating data
np.random.seed(42)
X = 2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.randn(100,1)   ## y = 4 + 3x + noise 
```

```python
# visualizing the dataset
plt.scatter(X, y, color= 'blue', label= 'Data points')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Generated data')
plt.legend()
plt.show()
```

```python
class LinearRegressionScratch:
    def __init__(self, learning_rate=0.1, iterations = 1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.theta = None
    
    def compute_cost(self,X,y):
        m = len(y)
        predictions = X.dot(self.theta)
        cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
        return cost
    
    def fit(self, X, y):
        m = len(y)
        X_b = np.c_[np.ones((m, 1)), X]  # Adding bias term
        self.theta = np.random.randn(2,1)  ## Initializing theta randomly

        self.cost_history = []


        for i in range(self.iterations):
            gradients = (1 / m) * X_b.T.dot(X_b.dot(self.theta) - y)
            self.theta -= self.learning_rate * gradients
            cost = self.compute_cost(X_b, y)
            self.cost_history.append(cost)

    def predict(self, X):
        X_b = np.c_[np.ones((len(X), 1)), X]  # Add bias term
        return X_b.dot(self.theta)
```

```python
### Training the model 

lr = LinearRegressionScratch(learning_rate=0.1,iterations=1000)
lr.fit(X, y)
```

```python
### Plotting the cost function
plt.plot(range(len(lr.cost_history)), lr.cost_history, color='red')
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function over Iterations")
plt.show()
```

```python
### visualizing the linear regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, lr.predict(X), color='red', label='Regression Line')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```
